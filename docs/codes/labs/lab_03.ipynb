```python
df_a = df.loc[df['type'].str.contains('Movie', case=False)]
df_a.head()
```

```python
#5)b) Se mostrar todas las prliculas y los minutos que dura
df_b = df.loc[df['type'].str.contains('Movie', case=False)]
df_movie = df_b[['title']].copy()
df_movie['Min'] = df['duration'].str.extract(r'(\d+)').fillna(0).astype(int)
df_movie.head(100)
```

```python
df['listed_in'] = df['listed_in'].str.split(', ')
df_e = df['listed_in'].explode('listed_in')
df_e = df_e.drop_duplicates()
df_e.head(10)
```

```python
df_op = df.explode('listed_in')
df_unu = df_op['listed_in'].value_counts()
df_unu.head(10)
```

```python
df_movie['Largo o corto'] = df_movie['Min'].where( df_movie['Min'] <=120, 'Largo').mask( df_movie['Min'] <=120, 'Corto')

df_movie.head()
```

```python
df_juju = df.loc[df['type'].str.contains('Movie', case=False)].copy()
df_juju['duration'] = df['duration'].str.extract(r'(\d+)').fillna(0).astype(int)
df_juju = df_juju.loc[(df_juju['duration']>=100)&(df_juju['rating']=='R')&(df_juju['country']=='United States')]
df_juju.head(100)
```

```python
df_lol = df.loc[df['type'].str.contains('Movie', case=False)].copy()
df_lol['duration'] = df['duration'].str.extract(r'(\d+)').fillna(0).astype(int)
df_lol = df_lol.sort_values('duration', ascending = False)
df_lol.head(10).style
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')
df.head()
```

## 📊 Parte 1: Exploración visual básica

1. **Distribución de tipos de contenido**

   * Realiza un gráfico de barras mostrando la cantidad de películas vs series (`type`).
   * Pregunta guía: ¿Cuál es el tipo de contenido predominante en Netflix? ¿Qué implicancias podría tener para los usuarios?

2. **Histograma de años de lanzamiento**

   * Muestra cómo se distribuyen los títulos según su `release_year`.
   * Pregunta guía: ¿Hay más contenido reciente o antiguo? ¿Notas algún patrón en décadas específicas?

3. **Proporción de clasificaciones por edad (`rating`)**

   * Haz un gráfico circular o de barras.
   * Pregunta guía: ¿Qué clasificación es más común? ¿Crees que Netflix está más enfocado en público adulto, familiar o infantil?



```python
Can_peli_serie=df["type"].value_counts()
x= Can_peli_serie.index
y= Can_peli_serie.values
maxcan= Can_peli_serie.max()
maxtipo= Can_peli_serie[Can_peli_serie == maxcan].index[0]
plt.bar(x, y)
plt.xlabel('Tipo de contenidos')
plt.ylabel('Cantidad')
plt.title('Tipos de contenidos en netflix')
plt.show()
print(f"El tipo que predomina en netflix son las {maxtipo} con {maxcan}, la implicancia que podria tener esto a los usarios es que es un consumo mas rapido, es decir busca a gente que prefiere visualizacaiones rapidas y no de largo consumos como las series")
```

```python
plt.hist(data = df, x="release_year", bins=30, edgecolor='black')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Cantidad de titulos')
plt.title('Histograma de los titulos segun año de lanzamiento')
plt.show()
print("Hay mas contenidos reciente, hay un patron de crecimiento muy elevado desde los años 2000, antes hay una presencia casi nula antes del 1950 y salieron pocas peliculas entre 1950 y 2000")
```

```python
Canrating=df["rating"].value_counts()
x= Canrating.index
y= Canrating.values
maxcant= Canrating.max()
maxtipo= Canrating[Canrating == maxcant].index[0]
plt.figure(figsize=(14,6))
plt.bar(x, y)
plt.xlabel('Clasificacion ')
plt.ylabel('Cantidad')
plt.title('Clasificacion por edad en Netflix ')
plt.show()
print(f"La clasificacion mas frecuente es {maxtipo} con una cantidad de {maxcant}, esta mas enfocado en adolecentes y adultos jovenes")
```



## 🎨 Parte 2: Tendencias y evolución en el tiempo

4. **Número de títulos agregados por año**

   * Usa `date_added` (convertido a fecha) y grafica una serie temporal.
   * Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?

5. **Heatmap de lanzamientos por año y mes**

   * Construye un mapa de calor con `year_added` y `month_added`.
   * Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

6. **Duración de películas por género**

   * Extrae la duración en minutos y crea un boxplot agrupado por género principal.
   * Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?


```python
df["date_added"] = df["date_added"].astype(str).str.strip()
df["date_added"]= pd.to_datetime(df["date_added"])
df["year_added"] =df["date_added"].dt.year
titanio =df["year_added"].value_counts().sort_index()
plt.plot(titanio.index, titanio.values, color='blue')
plt.title('Numero de titulos agregados a Netflix por año')
plt.xlabel('Año')
plt.ylabel('Numero de titulos')
plt.show()
print("El mayor aumento de catalogo fue entre los años 2016 y 2019 y que relacion, talvez la expasion quisieron tener mas diversidad y ampliar sus catalogos para incorporarse en otros paices ")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
df["year_added"]= df["date_added"].dt.year
df["month_added"]= df["date_added"].dt.month
heatdat =  (df.groupby(["month_added", "year_added"]).size().unstack(fill_value=0).reindex(index=range(1,13), fill_value=0))
plt.figure(figsize=(12,6))
im = plt.imshow(heatdat.values, cmap='Blues', aspect='auto', origin='upper',interpolation='nearest')
plt.colorbar(im, label="Cantidad de lanzamientos")
years = heatdat.columns.tolist()
plt.xticks(range(len(years)),[str(y) for y in years], rotation=45,ha="right")
plt.yticks(range(12), ["Enero","Febrero","Marzo", "Abril","Mayo","Junio","julio","agosto","septiembre","octubre","Noviembre","Diciembre"])
plt.xlabel("Año de lanzamiento")
plt.ylabel("Mes de lanzamiento")
plt.title("Heatmap de lanzamientos por año y mes")
plt.tight_layout()
plt.show()
```

```python
df_movie= df[df["type"] == "Movie"].copy()
df_movie["duration"] = df_movie["duration"].str.replace(" min", "").astype(float)
df_movie["main_genre"] = df_movie["listed_in"].str.split(",").str[0].str.strip()
plt.figure(figsize=(12,6))
sns.boxplot(data=df_movie, x="main_genre", y="duration")
plt.xticks(rotation=90)
plt.xlabel("Genero principal")
plt.ylabel("Duracion (min)")
plt.title("Duracion de peliculas por genero")
plt.show()
```

   * Haz un gráfico de dispersión (scatterplot) con `release_year` en el eje X y `duration` (en minutos) en el eje Y.

```python
top_countries =df["country"].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, color="Blue")
plt.xlabel("Número de producciones")
plt.ylabel("País")
plt.title("Top 10 países con más producciones en el catálogo")
plt.show()
print("Podemos ver por el grafico que domina Estados unidos y despues India, se eeseeperaba que fuera estados unidos por que esta hollywood ")
```

```python
df["main_genre"]= df["listed_in"].str.split(",").str[0].str.strip()
genre_type_counts =df.groupby(["main_genre", "type"]).size().unstack(fill_value=0)
genre_type_counts.plot(kind="bar",stacked=True,figsize=(12,6),color=["Blue","green"])
plt.xlabel("Genero principal")
plt.ylabel("Cantidad de titulos")
plt.title("Peliculas vs Series segun genero")
plt.xticks(rotation=90)
plt.legend(title="Tipo")
plt.tight_layout()
plt.show()
print("si, los generos de drama y action y adventure predominan en las pelis y en las series son el international y shows  ")
```

```python
df_movies= df[df["type"] == "Movie"].copy()
df_movies["duration"] =df_movies["duration"].str.replace(" min", "").astype(float)
plt.figure(figsize=(12,6))
sns.scatterplot(data=df_movies, x="release_year", y="duration", alpha=0.5)
plt.xlabel("Año de lanzamiento")
plt.ylabel("Duracion (min)")
plt.title("Relacion entre duracion y año de lanzamiento")
plt.show()
print("los estanderes de duracion se mantiene entre 90 y 120 minutos a lo largo de los años y si se ven unos outliers, con peliculas muy largas como unos 300 minutos y peliculas muy cortas ")
```

## ⭐ Desafío Final

* Encuentra las combinaciones más frecuentes de **género + rating** y represéntalas en un **heatmap o gráfico de burbujas**.
* Pregunta guía: ¿Qué géneros están más dirigidos a un público adulto? ¿Y cuáles a público familiar o infantil?

```python
df["main_genre"]= df["listed_in"].str.split(",").str[0].str.strip()
genre_rating = df.groupby(["main_genre", "rating"]).size().unstack(fill_value=0)
plt.figure(figsize=(12,6))
sns.heatmap(genre_rating, cmap="Blues", annot=True, fmt="d", cbar_kws={'label': 'Cantidad'})
plt.title("frecuencia de combinaciones genero ¡ rating")
plt.xlabel("rating")
plt.ylabel("Genero principal")
plt.show()
print("los generos esta mas dirigidos a adultos de dramas, comedias, international y clasificaciones como Tv-ma y R")
```

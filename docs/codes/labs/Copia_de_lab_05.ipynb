```python
Can_peli_serie=df["type"].value_counts()
x= Can_peli_serie.index
y= Can_peli_serie.values
maxcan= Can_peli_serie.max()
maxtipo= Can_peli_serie[Can_peli_serie == maxcan].index[0]
plt.bar(x, y)
plt.xlabel('Tipo de contenidos')
plt.ylabel('Cantidad')
plt.title('Tipos de contenidos en netflix')
plt.show()
print(f"El tipo que predomina en netflix son las {maxtipo} con {maxcan}, la implicancia que podria tener esto a los usarios es que es un consumo mas rapido, es decir busca a gente que prefiere visualizacaiones rapidas y no de largo consumos como las series")
```

```python
Canrating=df["rating"].value_counts()
x= Canrating.index
y= Canrating.values
maxcant= Canrating.max()
maxtipo= Canrating[Canrating == maxcant].index[0]
plt.figure(figsize=(14,6))
plt.bar(x, y)
plt.xlabel('Clasificacion ')
plt.ylabel('Cantidad')
plt.title('Clasificacion por edad en Netflix ')
plt.show()
print(f"La clasificacion mas frecuente es {maxtipo} con una cantidad de {maxcant}, esta mas enfocado en adolecentes y adultos jovenes")
```

6. **Duraci칩n de pel칤culas por g칠nero**

   * Extrae la duraci칩n en minutos y crea un boxplot agrupado por g칠nero principal.
   * Pregunta gu칤a: 쯈u칠 g칠neros suelen durar m치s? 쯈u칠 podr칤as decir sobre la variabilidad dentro de un mismo g칠nero?


```python
df["date_added"] = df["date_added"].astype(str).str.strip()
df["date_added"]= pd.to_datetime(df["date_added"])
df["year_added"] =df["date_added"].dt.year
titanio =df["year_added"].value_counts().sort_index()
plt.plot(titanio.index, titanio.values, color='blue')
plt.title('Numero de titulos agregados a Netflix por a침o')
plt.xlabel('A침o')
plt.ylabel('Numero de titulos')
plt.show()
print("El mayor aumento de catalogo fue entre los a침os 2016 y 2019 y que relacion, talvez la expasion quisieron tener mas diversidad y ampliar sus catalogos para incorporarse en otros paices ")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
df["year_added"]= df["date_added"].dt.year
df["month_added"]= df["date_added"].dt.month
heatdat =  (df.groupby(["month_added", "year_added"]).size().unstack(fill_value=0).reindex(index=range(1,13), fill_value=0))
plt.figure(figsize=(12,6))
im = plt.imshow(heatdat.values, cmap='Blues', aspect='auto', origin='upper',interpolation='nearest')
plt.colorbar(im, label="Cantidad de lanzamientos")
years = heatdat.columns.tolist()
plt.xticks(range(len(years)),[str(y) for y in years], rotation=45,ha="right")
plt.yticks(range(12), ["Enero","Febrero","Marzo", "Abril","Mayo","Junio","julio","agosto","septiembre","octubre","Noviembre","Diciembre"])
plt.xlabel("A침o de lanzamiento")
plt.ylabel("Mes de lanzamiento")
plt.title("Heatmap de lanzamientos por a침o y mes")
plt.tight_layout()
plt.show()
```

```python
df_movie= df[df["type"] == "Movie"].copy()
df_movie["duration"] = df_movie["duration"].str.replace(" min", "").astype(float)
df_movie["main_genre"] = df_movie["listed_in"].str.split(",").str[0].str.strip()
plt.figure(figsize=(12,6))
sns.boxplot(data=df_movie, x="main_genre", y="duration")
plt.xticks(rotation=90)
plt.xlabel("Genero principal")
plt.ylabel("Duracion (min)")
plt.title("Duracion de peliculas por genero")
plt.show()
```

```python
top_countries =df["country"].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, color="Blue")
plt.xlabel("N칰mero de producciones")
plt.ylabel("Pa칤s")
plt.title("Top 10 pa칤ses con m치s producciones en el cat치logo")
plt.show()
print("Podemos ver por el grafico que domina Estados unidos y despues India, se eeseeperaba que fuera estados unidos por que esta hollywood ")
```

```python
df["main_genre"]= df["listed_in"].str.split(",").str[0].str.strip()
genre_type_counts =df.groupby(["main_genre", "type"]).size().unstack(fill_value=0)
genre_type_counts.plot(kind="bar",stacked=True,figsize=(12,6),color=["Blue","green"])
plt.xlabel("Genero principal")
plt.ylabel("Cantidad de titulos")
plt.title("Peliculas vs Series segun genero")
plt.xticks(rotation=90)
plt.legend(title="Tipo")
plt.tight_layout()
plt.show()
print("si, los generos de drama y action y adventure predominan en las pelis y en las series son el international y shows  ")
```

```python
df_movies= df[df["type"] == "Movie"].copy()
df_movies["duration"] =df_movies["duration"].str.replace(" min", "").astype(float)
plt.figure(figsize=(12,6))
sns.scatterplot(data=df_movies, x="release_year", y="duration", alpha=0.5)
plt.xlabel("A침o de lanzamiento")
plt.ylabel("Duracion (min)")
plt.title("Relacion entre duracion y a침o de lanzamiento")
plt.show()
print("los estanderes de duracion se mantiene entre 90 y 120 minutos a lo largo de los a침os y si se ven unos outliers, con peliculas muy largas como unos 300 minutos y peliculas muy cortas ")
```
